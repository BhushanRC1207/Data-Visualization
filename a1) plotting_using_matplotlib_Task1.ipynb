### `Problem-9` Show Quarter wise Sales data for all products as multi Bar chart.


```python
pd.to_datetime(['2020-{}-01'.format(month) for month in df['month_number']])
```

```python
df['date'] = pd.to_datetime(['2020-{}-01'.format(month) for month in df['month_number']])
```

```python
final_df = df.groupby(df['date'].dt.quarter).sum()
```

```python
final_df
```

```python
i = -1
for col in final_df.columns[1:7]:
  plt.bar(final_df.index + i,final_df[col],width=0.15,label=col)
  i = i - 0.15

plt.xticks(final_df.index-1.4,final_df.index)
#plt.xlabel('Product')
#plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
final_df.iloc[:,1:7]
```

### `Problem-10` Plot Stacked Bar chart quarter wise for each product.

```python
all_cols = []

for col in final_df.columns[1:7]:
  if len(all_cols) == 0:
    plt.bar(final_df.index,final_df[col],label=col)
  else:
    plt.bar(final_df.index,final_df[col],bottom=sum(all_cols),label=col)
  all_cols.append(final_df[col])

plt.xticks(final_df.index - 0.02, final_df.index)
plt.legend()
plt.show()
```

```python
df=pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,label='2015')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,label='2016')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,label='2017')

plt.title('Batsman Runs Year Wise')
plt.xlabel('Batsman')
plt.ylabel('Runs')

plt.xticks(np.arange(df.shape[0]),df['batsman'],rotation='vertical')
plt.legend()
plt.show()
```

```python
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=(df['2015']+df['2016']),label='2017')

plt.legend()
```
