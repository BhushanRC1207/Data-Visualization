### **Problem-3:** Make a 3D *Surface* plot of this below mathematical equation.

$$z = |x| - |y|$$

````python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)

xx,yy=np.meshgrid(x,y)

z=np.abs(xx)-np.abs(yy)

fig=plt.figure(figsize=(10,8))
ax=plt.subplot(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

p=ax.plot_surface(xx,yy,z,cmap='viridis')
plt.colorbar(p)
plt.show()
````

### `Problem 9:` Create a surface plot and the 2 types of the contour plots of the below equation.

$$z = sin(\sqrt{x^2 + y^2})$$


````python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)

xx,yy=np.meshgrid(x,y)

z=np.sin(((xx**2+yy**2))**0.5)
````

````python
fig=plt.figure()

ax=plt.subplot(projection='3d')

p=ax.plot_surface(xx,yy,z,cmap='viridis')
plt.colorbar(p)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
````

````python
fig = plt.figure()
ax = plt.subplot()

ax.contour(xx,yy,z,cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
````

````python
fig = plt.figure()
ax = plt.subplot()

ax.contourf(xx,yy,z,cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
````

### `Problem 10:` Create a surface plot and the 2 types of the contour plots of the below equation.

$$z = tan(\log_2({x^2 + y^2})$$

````python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)

xx,yy=np.meshgrid(x,y)

z=np.tan(np.log2(xx**2+yy**2))
````

````python
fig = plt.figure()
ax = plt.subplot(projection='3d')

ax.plot_surface(xx,yy,z,cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
````

````python
fig = plt.figure()
ax = plt.subplot()

ax.contour(xx,yy,z,cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
````

````python
fig = plt.figure()
ax = plt.subplot()

ax.contourf(xx,yy,z,cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
````
